```python

from scipy import stats as st
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
x = df5[:, 1:-1]
y = df5[:, -1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y)

k = 3
y_test_pred = np.zeros(y_test.shape[0])

for l in range(x_test.shape[0]):
    dist = np.sqrt(np.sum((x_train - x_test[l])**2, axis=1))
    weight = 1 / (dist + 1e-20)**2
    maxkindices = np.zeros(k, dtype=int)
    weight1 = np.zeros(k, dtype=float)
    for i in range(k):
       maxkindices[i] = y_train[np.argmax(weight)]
       weight[np.argmax(weight)] = -np.inf
       
    y_test_pred[l] = st.mode(maxkindices)[0]
c = 0
for i in range(y_test.shape[0]):
    if y_test[i] == y_test_pred[i]:
        c += 1
accuracy = (c) / y_test.shape[0]
print("Accuracy:",accuracy)
precision = precision_score(y_test, y_test_pred, average='weighted')
recall = recall_score(y_test, y_test_pred, average='weighted')
f1 = f1_score(y_test, y_test_pred, average='weighted')
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error, mean_squared_error
import numpy as np

df6 = np.array(df6)

x = [[row[i] for i in range(5)] + [row[i] for i in range(6, 11)] for row in df6]
y = df6[:, 5]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

k = 3
pred = np.zeros(y_test.shape[0])

for l in range(len(x_test)):
    dist = np.sqrt(np.sum((np.array(x_train) - np.array(x_test[l]))**2, axis=1))
    weight = 1 / (dist + 1e-20)**2
    maxkindices = np.zeros(k, dtype=int)
    weight1 = np.zeros(k, dtype=float)
    for i in range(k):
        maxkindices[i] = y_train[np.argmax(weight)]
        weight1[i] = np.max(weight)
        weight[np.argmax(weight)] = -np.inf
    pred[l] = np.sum(maxkindices * weight1) / np.sum(weight1)


mape = mean_absolute_percentage_error(y_test, pred)
mae = mean_absolute_error(y_test, pred)
mse = mean_squared_error(y_test, pred)

print("Using inbuilt function")
print("MAE:", mae)
print("MSE:", mse)
print("MAPE:", mape)
```
